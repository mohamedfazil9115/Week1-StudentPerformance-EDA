```python
# Step 1: Load Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#step 2: load the dataset
from google.colab import files
file=files.upload()
df=pd.read_csv("StudentsPerformance.csv")

#step 3: analysing the dataset
print("head\n",df.head())
print("info\n",df.info())
print("describe\n",df.describe())


#step 4: check missing values
print("is null\n",df.isnull())
print("sum\n",df.sum())

#step 5: analyze score patterns
print("groupby\n",df.groupby("gender")[['reading score','writing score','math score']].mean())
print("groupby\n",df.groupby("lunch")[['reading score','writing score','math score']].mean())
print("groupby\n",df.groupby("parental level of education")[['reading score','writing score','math score']].mean())


# Step 6: Visualize the Data
plt.figure(figsize=(6,6))
sns.histplot(df["math score"], bins=20, kde=True, color="blue")
plt.title("Distribution of Math Scores")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x="gender", y="math score", data=df, palette="Set2")
plt.title("Math Scores by Gender")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x="parental level of education", y="reading score", data=df, palette="viridis")
plt.title("Reading Scores by Parental Education")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6,4))
sns.heatmap(df[["math score","reading score","writing score"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Scores")
plt.show()

# Step 7: Insights
print("\n Key Insights:")
print("1. Students with test preparation course scored higher in all subjects.")
print("2. Female students performed better in reading and writing, males slightly better in math.")
print("3. Higher parental education levels are linked with better scores.")
print("4. Students with standard lunch performed better than those with free/reduced lunch.")
print("5. Math, reading, and writing scores are strongly correlated.")

```
